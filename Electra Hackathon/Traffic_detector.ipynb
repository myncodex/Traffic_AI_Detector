```python
import cv2
import matplotlib.pyplot as plt
import cvlib as cv
import webbrowser
import time
from cvlib.object_detection import draw_bbox
cap= cv2.VideoCapture('video.mp4')
cap.set(cv2.CAP_PROP_FPS,10)
while True:
    ret,frame1=cap.read()
    bbox, label, conf = cv.detect_common_objects(frame1)
    output_image = draw_bbox(frame1, bbox, label, conf)
    #print('Number of cars in the image is '+ str(label.count('car')))
    if label.count('car')>15:
        message="Traffic_Ahead"
    else:
        message="Safe_Drive"
    print(message)
    html_content = f"<!DOCTYPE html><html lang = \"en\"> <head><title>Traffic cogestion </title><link rel=\"stylesheet\" href=\"style.css\"> </head> <body> <h2 data-text=\"{message}\">{message}</h2> </body> </html>"
    with open("index.html","w") as html_file:
        html_file.write(html_content)
    cv2.imshow('Video Orignal',frame1)
    if  cv2.waitKey(2)&0xFF==ord('q'):
        break
cv2.destroyAllWindows()
cap.release()
```
